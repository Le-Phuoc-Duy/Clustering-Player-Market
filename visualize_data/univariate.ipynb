```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train_player = pd.read_csv('../clean_data_train/players.csv')
test_player = pd.read_csv('../clean_data_test/players.csv')  
```

### Vị trí
___

```python
plt.figure(figsize=(10, 5))

position_labels = []
for position in train_player['position'].value_counts().index:
    if position == 'G':
        position_labels.append('Thủ môn')
    elif position == 'D':
        position_labels.append('Hậu vệ')
    elif position == 'M':
        position_labels.append('Tiền vệ')
    elif position == 'F':
        position_labels.append('Tiền đạo')
plt.subplot(1, 2, 1)
plt.pie(train_player['position'].value_counts(), labels=position_labels, autopct='%1.1f%%',textprops={'fontsize': 11})
plt.title('Training data ',fontsize=22) 

position_labels = []
for position in test_player['position'].value_counts().index:
    if position == 'G':
        position_labels.append('Thủ môn')
    elif position == 'D':
        position_labels.append('Hậu vệ')
    elif position == 'M':
        position_labels.append('Tiền vệ')
    elif position == 'F':
        position_labels.append('Tiền đạo')
plt.subplot(1, 2, 2)
plt.pie(test_player['position'].value_counts(), labels=position_labels, autopct='%1.1f%%',textprops={'fontsize': 11}) 
plt.title('Test data',fontsize=22) 
plt.tight_layout()  
 
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
foot_labels = []
for foot in train_player['preferred_foot'].value_counts().index:
    if foot == 'Right':
        foot_labels.append('Phải')
    elif foot == 'Left':
        foot_labels.append('Trái')
    elif foot == 'Both':
        foot_labels.append('Cả hai chân')

plt.pie(train_player['preferred_foot'].value_counts().values, labels= foot_labels, autopct='%1.1f%%',textprops={'fontsize': 11})
plt.title('Training data',fontsize=22) 

foot_labels = []
for foot in test_player['preferred_foot'].value_counts().index:
    if foot == 'Right':
        foot_labels.append('Phải')
    elif foot == 'Left':
        foot_labels.append('Trái')
    elif foot == 'Both':
        foot_labels.append('Cả hai chân')
plt.subplot(1, 2, 2)
plt.pie(test_player['preferred_foot'].value_counts().values, labels= foot_labels, autopct='%1.1f%%',textprops={'fontsize': 11})
plt.title('Test data',fontsize=22) 
```
